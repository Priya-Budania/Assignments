```python
total=soup.find_all('div',class_='jobInfoCard')
total
```

```python
len(total)
```

```python
role=[]
company=[]
experience=[]
salary=[]
skills=[]

for i in total:
    role.append(i.find('a',class_='title').text)
    company.append(i.find('p',class_='company').text.strip())
    entities=i.find_all('div',class_='entity')

    if len(entities)>0:
        experience.append(entities[0].text.strip())
    else:
        experience.append("N/A")

    if len(entities)>3:
        salary.append(entities[1].text.strip())
    else:
        salary.append("N/A")

    if len(entities)>2:
        skills.append(entities[-1].text.strip().split('\n')[0])
    else:
        skills.append('N/A')


    df = pd.DataFrame({
        'Job':role,
        'Company':company,
        'Experience':experience,
        'Salary':salary,
        'Skills':skills
    })
    
```

```python
final=pd.DataFrame()

for j in range(0,80):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    response = requests.get('https://www.ambitionbox.com/jobs/search?tag=data%20science&page={}'.format(j),headers=headers).text
    soup=BeautifulSoup(response,'html.parser')

    total=soup.find_all('div',class_='jobInfoCard')
    role=[]
    company=[]
    experience=[]
    salary=[]
    skills=[]

    for i in total:
        role.append(i.find('a',class_='title').text)
        company.append(i.find('p',class_='company').text.strip())
        entities=i.find_all('div',class_='entity')

        if len(entities)>0:
            experience.append(entities[0].text.strip())
        else:
            experience.append("N/A")

        if len(entities)>3:
            salary.append(entities[1].text.strip())
        else:
            salary.append("N/A")

        if len(entities)>2:
            skills.append(entities[-1].text.strip().split('\n')[0])
        else:
            skills.append('N/A')


        dataset = pd.DataFrame({
            'Job':role,
            'Company':company,
            'Experience':experience,
            'Salary':salary,
            'Skills':skills
        })

        final=pd.concat([final,dataset],ignore_index=True)
```
